```python
# An analysis on calories burnt by activity type - 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ActivityCals = pd.read_csv('ActivityCalories.csv')
ActivityCals
```

```python
df = pd.DataFrame(ActivityCals)

# Group the dataframe by the 'Day' and 'Activity Type' columns, then sum the values and unstack.
grouped = df.groupby(['Day','Activity Type']).sum()['Value'].unstack()
print(grouped)

# Reorder bars
order = ['Tue','Wed','Thu','Fri','Sat']

# Create a stacked bar chart from the grouped data
ax = grouped.loc[order].plot(kind='bar', stacked=True)
plt.xlabel('Day')
plt.ylabel('Calories')
plt.title('Stacked Bar Chart Grouped by Day')
plt.legend(title='Activity Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Category': ['A', 'B', 'C', 'D', 'E'],
        'Value': [10, 5, 20, 15, 8]}

# Create a DataFrame
df = pd.DataFrame(data)

# Define the custom order of categories
custom_order = ['A', 'C', 'D', 'E', 'B']

# Reorder the DataFrame based on the custom order
df['Category'] = pd.Categorical(df['Category'], categories=custom_order, ordered=True)
df = df.sort_values('Category')

# Create the bar chart
plt.bar(df['Category'], df['Value'])
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Chart with Custom Order')

plt.show()
```
